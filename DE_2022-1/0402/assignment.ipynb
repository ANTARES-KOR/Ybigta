## 이제부터 셀의 빈 공간에 주어진 결과를 출력하는 코드를 작성해 주시면 됩니다.
### rdd에 transform 연산을 적용할 때 rdd 변수에 저장해서 덮어씌워주세요.
연산을 적용한 후 복원하지 않고 하나씩 계속 누적해서 적용할 것입니다!
### Action 연산은 따로 작성하지 않아주셔도 됩니다.
예시)   
rdd = rdd.map(~)   
rdd.collect()

```python
# 값이 ''인 레코드를 제거한 후 배열로 출력
rdd = rdd.filter(lambda x: x != '')
rdd.collect()
```

```python
# 각 레코드를 공백을 기준으로 분리 후 하나의 배열로 출력하기
# ['The', 'Project', 'Gutenberg'..]
rdd = rdd.flatMap(lambda x: x.split())
rdd.collect()
```

```python
# 각 레코드에 포함된 특수문자를 제거한 후
# 위 셀과 같은 형태로 출력하기
rdd = rdd.map(lambda x: re.sub(r"[^a-zA-Z0-9]", "", x))
rdd.collect()
```

```python
# 특수문자만 있던 단어가 삭제되어 생긴 '' 값 제거(위에서 ''를 제거한 연산과 동일)
rdd = rdd.filter(lambda x: x != '')
```

```python
# 단어들을 모두 소문자로 변경한 후 배열로 출력
rdd = rdd.map(lambda x: x.lower())
rdd.collect()
```

```python
# 불용어 처리
# en_stops에 있는 단어만 배열로 출력
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
en_stops = set(stopwords.words('english'))

rdd = rdd.filter(lambda x: x not in en_stops)


rdd.collect()
```

```python
# key-value rdd로 변환한 후 배열로 출력
# (각 단어, 단어 길이) 형식
rdd = rdd.map(lambda x: (x, len(x)))
rdd.collect()
```

```python
# key를 기준으로 value 값을 모두 더하는 연산을 한 뒤 배열로 출력
rdd = rdd.reduceByKey(lambda x, y: x+y)
rdd.collect()
```

```python
# value값을 기준으로 내림차순 정렬한 후 배열로 출력
rdd = rdd.sortBy(lambda x: x[1], False)
rdd.collect()
```
